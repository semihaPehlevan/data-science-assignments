```python
# İKİ SAYININ EN BÜYÜK ORTAK BÖLENİ(EBOB)

def Ebob(a, b):
    while b:
        temp = a % b
        a = b
        b = temp
    return a

print(Ebob(7, 13))
```

```python
def anagram_sort(kelime1, kelime2):
    # Normalizasyon: boşlukları kaldır, küçük harfe çevir
    kelime1 = ''.join(harf.lower() for harf in kelime1 if harf.isalnum())
    kelime2 = ''.join(harf.lower() for harf in kelime2 if harf.isalnum())

    if len(kelime1) != len(kelime2):
        return False

    return sorted(kelime1) == sorted(kelime2)

print(anagram_sort("Listen", "Silent"))
```

```python
# LİSTEDEKİ EKSİK SAYIYI BULMA
import random

sonsayi = int(input("Liste eleman sayısını giriniz: "))
liste = list(range(1, sonsayi+1))

eleman = random.choice(liste)
liste.remove(eleman)
print(liste)

def eksikSayiyiBul(liste):
  n = len(liste) + 1
  toplam = n * (n + 1) // 2
  eksik = toplam - sum(liste)
  return eksik


print("Eksik sayı:", eksikSayiyiBul(liste))
```

```python
# LİSTEYİ BELİRLİ BİR DEĞERE GÖRE DÖNDÜRME

def liste_dondur(liste, k):
    n = len(liste)
    k = k % n
    return liste[-k:] + liste[:-k]

print(liste_dondur([1, 2, 3, 4, 5], 2))
```

```python
# İKİ SIRALI LİSTENİN MEDYANINI BULMA

def medyan_bulma(liste1, liste2):

    if len(liste1) > len(liste2):
        liste1, liste2 = liste2, liste1

    uzunluk1, uzunluk2 = len(liste1), len(liste2)
    alt, ust = 0, uzunluk1

    while alt <= ust:
        bolum1 = (alt + ust) // 2
        bolum2 = (uzunluk1 + uzunluk2 + 1) // 2 - bolum1

        solMax1 = float('-inf') if bolum1 == 0 else liste1[bolum1 - 1]
        sagMin1 = float('inf') if bolum1 == uzunluk1 else liste1[bolum1]

        solMax2 = float('-inf') if bolum2 == 0 else liste2[bolum2 - 1]
        sagMin2 = float('inf') if bolum2 == uzunluk2 else liste2[bolum2]

        if solMax1 <= sagMin2 and solMax2 <= sagMin1:
            if (uzunluk1 + uzunluk2) % 2 == 0:
                return (max(solMax1, solMax2) + min(sagMin1, sagMin2)) / 2.0
            else:
                return max(solMax1, solMax2)
        elif solMax1 > sagMin2:
            ust = bolum1 - 1
        else:
            alt = bolum1 + 1

print(medyan_bulma([1,3], [2]))
print(medyan_bulma([1,2], [3,4]))
```

```python
# EN UZUN PALİNDROMİK ALT STRİNG

def en_uzun_palindromik_altstring(metin):
    if not metin:
        return ""

    baslangic, bitis = 0, 0

    def genislet_merkezden(sol, sag):
        while sol >= 0 and sag < len(metin) and metin[sol] == metin[sag]:
            sol -= 1
            sag += 1
        return sag - sol - 1

    for i in range(len(metin)):
        uzunluk1 = genislet_merkezden(i, i)
        uzunluk2 = genislet_merkezden(i, i + 1)
        max_uzunluk = max(uzunluk1, uzunluk2)

        if max_uzunluk > (bitis - baslangic):
            baslangic = i - (max_uzunluk - 1) // 2
            bitis = i + max_uzunluk // 2

    return metin[baslangic:bitis + 1]

print(en_uzun_palindromik_altstring("babad"))
print(en_uzun_palindromik_altstring("cbbd"))
```

```python
# KELİME MERDİVENİ

from collections import deque

def kelime_merdiveni(baslangic, bitis, kelimeler):
    if bitis not in kelimeler:
        return 0

    kelime_kumesi = set(kelimeler)
    kuyruk = deque([(baslangic, 1)])

    while kuyruk:
        kelime, adim = kuyruk.popleft()

        for i in range(len(kelime)):
            for ch in "abcdefghijklmnopqrstuvwxyz":
                yeni_kelime = kelime[:i] + ch + kelime[i+1:]

                if yeni_kelime == bitis:
                    return adim + 1

                if yeni_kelime in kelime_kumesi:
                    kelime_kumesi.remove(yeni_kelime)
                    kuyruk.append((yeni_kelime, adim + 1))

    return 0

print(kelime_merdiveni("hit", "cog", ["hot","dot","dog","lot","log","cog"]))

print(kelime_merdiveni("hit", "cog", ["hot","dot","dog","lot","log"]))
```

```python
# MAKSİMUM ALT DİZİ TOPLAMI

def maksimum_alt_dizi_toplami(liste):
    mevcut_toplam = liste[0]
    en_yuksek_toplam = liste[0]

    for sayi in liste[1:]:
        mevcut_toplam = max(sayi, mevcut_toplam + sayi)

        en_yuksek_toplam = max(en_yuksek_toplam, mevcut_toplam)

    return en_yuksek_toplam

print(maksimum_alt_dizi_toplami([-2, 1, -3, 4, -1, 2, 1, -5, 4]))
print(maksimum_alt_dizi_toplami([5, 4, -1, 7, 8]))
```

```python
# BİRLEŞTİRME ARALIKLARI

def araliklari_birlestir(araliklar):
    if not araliklar:
        return []

    araliklar.sort(key=lambda x: x[0])

    birlesmis = [araliklar[0]]

    for baslangic, bitis in araliklar[1:]:
        son_birlesmis_bas, son_birlesmis_bit = birlesmis[-1]

        if baslangic <= son_birlesmis_bit:
            birlesmis[-1][1] = max(son_birlesmis_bit, bitis)
        else:
            birlesmis.append([baslangic, bitis])

    return birlesmis

print(araliklari_birlestir([[1,3],[2,6],[8,10],[15,18]]))
print(araliklari_birlestir([[1,4],[4,5]]))
```

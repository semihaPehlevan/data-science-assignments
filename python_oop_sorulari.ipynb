```python
# ÖĞRENCİ SINIFI

class Ogrenci:
    def __init__(self, isim, numara, notlar):
        self.isim = isim
        self.numara = numara
        self.notlar = notlar if notlar else []
        self.ortalama = self.ortalama_hesapla()

    def not_ekle(self, n):
        if 0 <= n <= 100:
            self.notlar.append(n)
            self.ortalama = self.ortalama_hesapla()
        else:
            print(f"Notlar 0-100 arası olmalıdır.")

    def ortalama_hesapla(self):
        if self.notlar:
            return round(sum(self.notlar)/len(self.notlar), 2)
        return 0

    def bilgileri_goster(self):
        print(f"Öğrenci: {self.isim} ({self.numara}), Notlar: {self.notlar}, Ortalama: {self.ortalama}")

ogr1 = Ogrenci("Semiha", "12345", [85, 90, 78])
ogr1.bilgileri_goster()

ogr1.not_ekle(95)
ogr1.bilgileri_goster()
ogr1.not_ekle(150)
```

```python
# HESAP MAKİNESİ SINIFI

class HesapMakinesi:
    def __init__(self):
        self.bellek = 0

    def topla(self, a, b):
        self.bellek = a + b
        return self.bellek

    def cikar(self, a, b):
        self.bellek = a - b
        return self.bellek

    def carp(self, a, b):
        self.bellek = a * b
        return self.bellek

    def bol(self, a, b):
        if b == 0:
            return "Sıfıra bölme mümkün değil."
        self.bellek = a / b
        return self.bellek

calc = HesapMakinesi()
print(calc.topla(10, 5))  # Çıktı: 15
print(calc.cikar(10, 5))  # Çıktı: 5
print(calc.carp(10, 5))   # Çıktı: 50
print(calc.bol(10, 5))    # Çıktı: 2.0
print(calc.bol(10, 0))    # Çıktı: Hata: Sıfıra bölme mümkün değil!
```

```python
# KİTAP SINIFI

class Kitap:
    def __init__(self,okuyan ,baslik, yazar, sayfa_sayisi):
        self.okuyan = okuyan
        self.baslik = baslik
        self.yazar = yazar
        if sayfa_sayisi < 0:
            raise ValueError("Sayfa sayısı negatif olamaz.")
        self.sayfa_sayisi = sayfa_sayisi
        self.okunan_sayfa = 0

    def sayfa_okundu(self, n):
        if n < 0:
            print("Negatif sayıda sayfa okunamaz.")
            return
        if self.okunan_sayfa + n > self.sayfa_sayisi:
            self.okunan_sayfa = self.sayfa_sayisi
        else:
            self.okunan_sayfa += n

    def kitap_bilgisi(self):
        print(f"{self.okuyan} - Yazar: {self.yazar} Kitap: {self.baslik} Okunan Sayfa Sayısı: {self.okunan_sayfa} ")
kitap1 = Kitap("Semiha Pehlevan","Ursula Le Guin","Rüyanın Öte Yakası", 300)
kitap1.kitap_bilgisi()

kitap1.sayfa_okundu(50)
kitap1.kitap_bilgisi()

kitap1.sayfa_okundu(300)
kitap1.kitap_bilgisi()
```

```python
# BANKA HESABI SINIFI

class BankaHesabi:
    def __init__(self, hesap_no, ad, bakiye):
        self.hesap_no = hesap_no
        self.ad = ad
        self.bakiye = bakiye
        self.islem_gecmisi = []

    def para_yatir(self, tutar):
        if tutar <= 0:
            print("Yatırılacak tutar pozitif olmalı.")
            return
        self.bakiye += tutar
        self.islem_gecmisi.append(f"+{tutar} TL")
        print(f"{tutar} TL yatırıldı. Güncel bakiye: {self.bakiye} TL")

    def para_cek(self, tutar):
        if tutar <= 0:
            print("Çekilecek tutar pozitif olmalı.")
            return
        if tutar > self.bakiye:
            print("Yetersiz bakiye.")
            return
        self.bakiye -= tutar
        self.islem_gecmisi.append(f"-{tutar} TL")
        print(f"{tutar} TL çekildi. Güncel bakiye: {self.bakiye} TL")

    def bakiye_sorgula(self):
        son_islem = self.islem_gecmisi[-1] if self.islem_gecmisi else "Yok"
        print(f"Bakiye: {self.bakiye} TL, Son işlem: {son_islem}")

hesap = BankaHesabi("12345", "Zeynep Lale Tepatam", 1000)
hesap.para_yatir(500)
hesap.para_cek(300)
hesap.bakiye_sorgula()
```

```python
# GEOMETRİK ŞEKİL SINIFI

import math

class Sekil:
    def alan(self):
        pass

    def cevre(self):
        pass

class Dikdortgen(Sekil):
    def __init__(self, genislik, yukseklik):
        if genislik < 0 or yukseklik < 0:
            raise ValueError("Boyutlar negatif olamaz.")
        self.genislik = genislik
        self.yukseklik = yukseklik

    def alan(self):
        return self.genislik * self.yukseklik

    def cevre(self):
        return 2 * (self.genislik + self.yukseklik)

class Daire(Sekil):
    def __init__(self, yaricap):
        if yaricap < 0:
            raise ValueError("Yarıçap negatif olamaz.")
        self.yaricap = yaricap

    def alan(self):
        return round(math.pi * self.yaricap ** 2, 2)

    def cevre(self):
        return round(2 * math.pi * self.yaricap, 2)

dik = Dikdortgen(5, 3)
print(f"Alan: {dik.alan()}, Çevre: {dik.cevre()}")

daire = Daire(4)
print(f"Alan: {daire.alan()}, Çevre: {daire.cevre()}")
```

```python
# ÇALIŞAN YÖNETİM SİSTEMİ

class Calisan:
    def __init__(self, isim, maas, departman):
        self.isim = isim
        self.maas = maas
        self.departman = departman

    def bilgi_goster(self):
        return f"{self.isim} - {self.departman} Departmanı, Maaş: {self.maas} TL"

class NormalCalisan(Calisan):
    pass

class Yonetici(Calisan):
    def __init__(self, isim, maas, departman, ekip_buyuklugu):
        super().__init__(isim, maas, departman)
        self.ekip_buyuklugu = ekip_buyuklugu

    def bilgi_goster(self):
        maas_bonus = self.maas + 1000 * self.ekip_buyuklugu
        return (f"{self.isim} - {self.departman} Departmanı, Yönetici, "
                f"Ekip: {self.ekip_buyuklugu} kişi, Maaş: {maas_bonus} TL")

class Muhendis(Calisan):
    def __init__(self, isim, maas, departman, proje_sayisi):
        super().__init__(isim, maas, departman)
        self.proje_sayisi = proje_sayisi

    def bilgi_goster(self):
        maas_bonus = self.maas + 850 * self.proje_sayisi
        return (f"{self.isim} - {self.departman} Departmanı, Mühendis, "
                f"Projeler: {self.proje_sayisi}, Maaş: {maas_bonus} TL")

calisan1 = NormalCalisan("Ali", 5000, "IT")
print(calisan1.bilgi_goster())

yonetici1 = Yonetici("Mehmet", 8000, "IT", 5)
print(yonetici1.bilgi_goster())

muhendis1 = Muhendis("Ayşe", 7000, "IT", 3)
print(muhendis1.bilgi_goster())
```

```python
# KÜTÜPHANE YÖNETİMİ

class Kitap:
    def __init__(self, baslik, yazar):
        self.baslik = baslik
        self.yazar = yazar
        self.odunc_alan = None

    def odunc_ver(self, uye):
        if self.odunc_alan is not None:
            print(f"{self.baslik} zaten {self.odunc_alan.isim}'de.")
            return False
        self.odunc_alan = uye
        return True

    def geri_al(self):
        self.odunc_alan = None

class Uye:
    def __init__(self, isim):
        self.isim = isim
        self.alinan_kitaplar = []

    def kitap_al(self, kitap):
        if len(self.alinan_kitaplar) >= 3:
            print(f"{self.isim} maksimum 3 kitap alabilir.")
            return False
        if kitap.odunc_ver(self):
            self.alinan_kitaplar.append(kitap)
            print(f"{kitap.baslik} kitabı {self.isim}'e verildi.")
            return True
        return False

    def kitap_iade(self, kitap):
        if kitap in self.alinan_kitaplar:
            kitap.geri_al()
            self.alinan_kitaplar.remove(kitap)
            print(f"{kitap.baslik} kitabı {self.isim} tarafından iade edildi.")

class Kutuphane:
    def __init__(self):
        self.kitaplar = []
        self.uyeler = []

    def kitap_ekle(self, kitap):
        self.kitaplar.append(kitap)

    def uye_ekle(self, uye):
        self.uyeler.append(uye)

    def durum(self):
        kütüphane_kitaplari = [k.baslik for k in self.kitaplar if k.odunc_alan is None]
        print(f"Kütüphanede kalan kitaplar: {len(kütüphane_kitaplari)}")
        for k in self.kitaplar:
            sahip = k.odunc_alan.isim if k.odunc_alan else "Kütüphanede"
            print(f"{k.baslik} - {sahip}")

kütüphane = Kutuphane()

kitap1 = Kitap("Ölülerle Uzlaşmak", "Margaret Atwood")
kitap2 = Kitap("Wendigo", "Algernon Blackwood")
kitap3 = Kitap("Gökteki Göz", "Philip K. Dick")
kitap4 = Kitap("Sonbahar Ülkesi", "Ray Bradbury")

kütüphane.kitap_ekle(kitap1)
kütüphane.kitap_ekle(kitap2)
kütüphane.kitap_ekle(kitap3)
kütüphane.kitap_ekle(kitap4)

uye1 = Uye("Semiha Pehlevan")
kütüphane.uye_ekle(uye1)

uye1.kitap_al(kitap1)
uye1.kitap_al(kitap2)
uye1.kitap_al(kitap3)
uye1.kitap_al(kitap4)
kütüphane.durum()
```

```python
# EV OTOMASYON SİSTEMİ

class Cihaz:
    def __init__(self, konum, durum=False):
        self.konum = konum
        self.durum = durum

    def ac(self):
        self.durum = True

    def kapat(self):
        self.durum = False

    def durum_bilgisi(self):
        return f"{self.konum} cihaz durumu: {'Açık' if self.durum else 'Kapalı'}"

class Isik(Cihaz):
    def durum_bilgisi(self):
        return f"{self.konum} ışığı {'açık' if self.durum else 'kapalı'}"

class Termostat(Cihaz):
    def __init__(self, konum, sicaklik=22):
        super().__init__(konum, True)
        self.sicaklik = sicaklik

    def ayarla(self, yeni_sicaklik):
        self.sicaklik = yeni_sicaklik

    def durum_bilgisi(self):
        return f"{self.konum} ısısı {self.sicaklik}°C"

class GuvenlikKamerasi(Cihaz):
    def __init__(self, konum, aktif=True):
        super().__init__(konum, aktif)

    def kayit_baslat(self):
        if self.durum:
            return f"{self.konum} kamerası kayıt başlattı"
        else:
            return f"{self.konum} kamerası kapalı"

    def durum_bilgisi(self):
        return f"{self.konum} kamerası {'aktif' if self.durum else 'pasif'}"

class EvSistemi:
    def __init__(self):
        self.cihazlar = []

    def cihaz_ekle(self, cihaz):
        self.cihazlar.append(cihaz)

    def tum_durumlar(self):
        for cihaz in self.cihazlar:
            print(cihaz.durum_bilgisi())

ev = EvSistemi()

isik1 = Isik("Salon", True)
term1 = Termostat("Oturma Odası", 22)
kamera1 = GuvenlikKamerasi("Giriş", True)

ev.cihaz_ekle(isik1)
ev.cihaz_ekle(term1)
ev.cihaz_ekle(kamera1)

ev.tum_durumlar()

```
